# 500 convert_responses_for_R

```python
import pandas as pd
import numpy as np
```

```python
responses_df = pd.read_excel('../data/model_responses/v2_all_response_df.xlsx', index_col=[0, 1])
responses_df
```

```python
col_list = []
for col in responses_df.columns:
    col_list.append(responses_df[col].explode())
    
frame_list = []
for series in col_list:
    temp_frame = series.to_frame().assign(question=series.name)
    temp_frame.rename(columns={temp_frame.columns[0]:'values'}, inplace=True)
    frame_list.append(temp_frame)
    
response_df_transformed = pd.concat(frame_list, axis=0)
response_df_transformed
```

```python
response_df_transformed['values'] = response_df_transformed['values'].apply(lambda x: eval(x)[1])
response_df_transformed
```

```python

for idx, row in response_df_transformed.iterrows():
    # Convert the dictionary in 'values' column to a DataFrame
    dict_df = pd.DataFrame(row['values'])

    # Explode the DataFrame if the dictionary values are lists
    dict_df = dict_df.explode(list(dict_df.columns))

    # Assigning the 'question' value and original index to the new DataFrame
    dict_df['question'] = row['question']
    dict_df = dict_df.set_index([pd.Index([idx]*len(dict_df), name=response_df_transformed.index.names), 'question'])

    # Concatenate with the original DataFrame if it's not the first iteration
    if 'response_scores_df' not in locals():
        response_scores_df = dict_df
    else:
        response_scores_df = pd.concat([response_scores_df, dict_df])
        
response_scores_df=response_scores_df.reset_index().rename(columns={'level_0':'affiliation', 'level_1':'model_size'}).set_index(['affiliation', 'model_size'])
response_scores_df['model'] = 'LLM'
response_scores_df
```

```python
human_scores_df = pd.read_json("../data/final_ft_datasets/scored_training_examples.json")
human_scores_df = human_scores_df.stack().to_frame()
human_scores_df.rename(columns={0:'values'}, inplace=True)
human_scores_df['question'] = 'comment'
human_scores_df
```

```python
for idx, row in human_scores_df.iterrows():
    # Convert the dictionary in 'values' column to a DataFrame
    dict_df = pd.DataFrame(row['values']['scores'])
    # Explode the DataFrame if the dictionary values are lists
    dict_df = dict_df.explode(list(dict_df.columns))

    # Assigning the 'question' value and original index to the new DataFrame
    dict_df['question'] = row['question']
    dict_df = dict_df.set_index([pd.Index([idx]*len(dict_df), name=human_scores_df.index.names), 'question'])

    # Concatenate with the original DataFrame if it's not the first iteration
    if idx == human_scores_df.first_valid_index():
        human_scores_transformed_df = dict_df
    else:
        human_scores_transformed_df = pd.concat([human_scores_transformed_df, dict_df])
        
human_scores_transformed_df=human_scores_transformed_df.reset_index().rename(columns={'level_1':'affiliation', 'level_0':'model_size'}).set_index(['affiliation', 'model_size'])
human_scores_transformed_df['model'] = 'human'
human_scores_transformed_df
```

```python
output_scores_df = pd.concat([response_scores_df, human_scores_transformed_df])
output_scores_df = output_scores_df.reset_index().rename(columns={'level_0':'affiliation', 'level_1':'model_size'})
output_scores_df
```

```python
output_scores_df.to_csv("../data/model_responses/regression_dataset_full.csv")
```
